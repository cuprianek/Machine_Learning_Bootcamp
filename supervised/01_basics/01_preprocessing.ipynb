### <a name='0'></a>Import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn
```

### <a name='1'></a>Wygenerowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}
df_raw = pd.DataFrame(data)
df_raw
```

### <a name='2'></a>Utworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

### <a name='0'></a>Zmiana typu danych i wstępna eksploracja

```python
for column in ['size', 'color', 'gender', 'bought']:
    df[column] = df[column].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include='category').T
```

### <a name='0'></a>LabelEncoder

głównie do zmiennych 0/1

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.fit_transform(df['bought'])
```

```python
le.classes_
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought'])
```

### <a name='0'></a>OneHotEncoder

do zamiany zmiennych opisowych

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop = 'first', sparse =False)
encoder.fit_transform(df[['size']])
```

### <a name='0'></a>Standaryzacja - StandardScaler

```python
print(f"{df['price']}")
print(f"Avg prive is {df['price'].mean()}")
print(f"Standard deviation is {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
def standarize(x):
    return (x - x.mean()) / x.std()

standarize(df['price'])
```

```python
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit_transform(df[['price']])
```

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price','weight']])
df
```

### <a name='0'></a>Przygotowanie danych do modelu

```python
df= df_raw.copy()
df
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder()
ss = StandardScaler()
df['bought'] = le.fit_transform(df['bought'])
df[['price', 'weight']] = ss.fit_transform(df[['price','weight']])
df = pd.get_dummies(df, drop_first=True, prefix_sep='_')


df
```
