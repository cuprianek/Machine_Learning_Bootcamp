### <a name='1'></a> Import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn
```

### <a name='1'></a> Załadowanie danych

```python
def fetch_financial_data(company='NKE'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df_raw = fetch_financial_data()
df_raw.head()
```

### <a name='1'></a> Utworzenie kopii danych

```python
df = df_raw.copy()
df = df[:5]
df
```

### <a name='1'></a> Generowanie nwoych zmiennych

```python
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df
```

### <a name='1'></a> Dyskretyzacja zmiennej ciągłej

```python
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins = 3)
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 185, 195))
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 185, 195), labels =['small', 'medium', 'tall'])
df
```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

### <a name='1'></a> Ekstrakcja cech

```python
df = pd.DataFrame(data={'lang': [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df
```

```python
df['lang_number'] = df['lang'].apply(len)
df
```

```python
df['lang_pl'] = df['lang'].apply(lambda x:1 if 'PL' in x else 0)
df
```
